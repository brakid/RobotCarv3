```python
# https://github.com/e-Gizmo/QMC5883L-GY-271-Compass-module/blob/master/QMC5883L%20Datasheet%201.0%20.pdf
import time
from compass import Compass
from compass_sensor import CompassSensor

compass = Compass()
compass_sensor = CompassSensor(compass)
compass_sensor.start()

i = 0
while True:
    try:
        heading = compass_sensor.get_heading()

        if i % 10 == 0:
            print(heading)
        i += 1

        if i == 1000:
            break
        
        time.sleep(0.01)
    except Exception as e:
        print(e, 'exiting...')
        break
```
